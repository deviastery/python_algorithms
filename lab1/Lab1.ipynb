```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
import random
import time

start_time = time.time()

Matrix = [ [ random.randint(1, 50) for j in range(10)] for i in range(4) ]

print(Matrix)

def selectionSort(arr):      

    for i in range(len(arr)):
        min = i
        
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min]:
                min = j

        arr[min], arr[i] = arr[i], arr[min]
            
    return arr

def matrixSelectionSort(matrix):

    sortMatrix = []
    rows = len(matrix)            
    cols = len(matrix[0])

    for row in range(rows):
        newRow = selectionSort(matrix[row])
        sortMatrix.append(newRow)
        
    return sortMatrix
 
print(matrixSelectionSort(Matrix))

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка вставкой.
import random
import time

start_time = time.time()

Matrix = [ [ random.randint(1, 50) for j in range(10)] for i in range(4) ]

print(Matrix)

def insertionSort(arr):
        
    for i in range(len(arr)):
        cursor = arr[i]
        pos = i
        
        while pos > 0 and arr[pos - 1] > cursor:
            arr[pos] = arr[pos - 1]
            pos = pos - 1
        
        arr[pos] = cursor

    return arr

def matrixInsertionSort(matrix):

    sortMatrix = []
    rows = len(matrix)            
    cols = len(matrix[0])

    for row in range(rows):
        newRow = insertionSort(matrix[row])
        sortMatrix.append(newRow)
        
    return sortMatrix
 
print(matrixInsertionSort(Matrix))

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка обменом. || Сортировка пузырьком.
import random
import time

start_time = time.time()

Matrix = [ [ random.randint(1, 50) for j in range(10)] for i in range(4) ]

print(Matrix)

def bubbleSort(arr):

    n = len(arr)
    swapped = True
    x = -1

    while swapped:
        swapped = False
        x = x + 1

        for i in range(1, n-x):
            if arr[i - 1] > arr[i]:
                arr[i - 1], arr[i] = arr[i], arr[i - 1]
                swapped = True
                
    return arr

def matrixBubbleSort(matrix):

    sortMatrix = []
    rows = len(matrix)            
    cols = len(matrix[0])

    for row in range(rows):
        newRow = bubbleSort(matrix[row])
        sortMatrix.append(newRow)
        
    return sortMatrix
 
print(matrixBubbleSort(Matrix))

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка Шелла.
import random
import time

start_time = time.time()

Matrix = [ [ random.randint(1, 50) for j in range(10)] for i in range(4) ]

print(Matrix)

def shellSort(arr):

    n = len(arr)
    interval = n // 2

    while interval > 0:
        for i in range(interval, n):
            temp = arr[i]
            j = i

            while j >= interval and arr[j - interval] > temp:
                arr[j] = arr[j - interval]
                j -= interval

            arr[j] = temp

        interval //= 2

    return arr

def matrixShellSort(matrix):

    sortMatrix = []
    rows = len(matrix)            
    cols = len(matrix[0])

    for row in range(rows):
        newRow = shellSort(matrix[row])
        sortMatrix.append(newRow)
        
    return sortMatrix
 
print(matrixShellSort(Matrix))

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Быстрая сортировка.
import random
import time

start_time = time.time()

Matrix = [ [ random.randint(1, 50) for j in range(10)] for i in range(4) ]

print(Matrix)

def quickSort(arr):

    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr) // 2]

    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return quickSort(left) + middle + quickSort(right)

def matrixQuickSort(matrix):

    sortMatrix = []
    rows = len(matrix)            
    cols = len(matrix[0])

    for row in range(rows):
        newRow = quickSort(matrix[row])
        sortMatrix.append(newRow)

    return sortMatrix
 
print(matrixQuickSort(Matrix))

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
    
```

```python
# Турнирная сортировка.
import random
import time

start_time = time.time()

Matrix = [[random.randint(1, 50) for j in range(10)] for i in range(4)]

print(Matrix)

def tournamentSort(arr):

    n = len(arr)
    tmp = [0] * (n * 2)
    INF = float('inf')

    def winner(pos1, pos2):
        u = pos1 if pos1 >= n else tmp[pos1]
        v = pos2 if pos2 >= n else tmp[pos2]

        if tmp[u] <= tmp[v]:
            return u
        
        return v

    def create_tree():
        for i in range(n):
            tmp[n + i] = arr[i]

        for i in range(2 * n - 1, 1, -2):
            k = i // 2
            j = i - 1
            tmp[k] = winner(i, j)

        value = tmp[tmp[1]]
        tmp[tmp[1]] = INF

        return value

    def recreate():
        i = tmp[1]

        while i > 1:
            j = k = i // 2

            if i % 2 == 0 and i < 2 * n - 1:
                j = i + 1
                
            else:
                j = i - 1

            tmp[k] = winner(i, j)
            i = k
            
        value = tmp[tmp[1]]
        tmp[tmp[1]] = INF

        return value

    
    arr[0] = create_tree()
    for i in range(1, n):
        arr[i] = recreate()

    return arr

def matrixTournamentSort(matrix):

    sortMatrix = []
    rows = len(matrix)
    cols = len(matrix[0])

    for row in range(rows):
        newRow = tournamentSort(matrix[row])
        sortMatrix.append(newRow)

    return sortMatrix

print(matrixTournamentSort(Matrix))

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```
